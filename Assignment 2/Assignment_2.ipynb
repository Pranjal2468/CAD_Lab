# Question 1

```julia
using Plots

x1 = 0.1:0.1:1
x2 = 0.1:0.1:1

x₁ = repeat(transpose(x1), 10)
x₂ = repeat(x2, 1, 10)

u1 = 0.2.*log.(1 .+x₁ .+x₂)
u2 = 0.2.*exp.(x₁)

x1_deformed = x₁ + u1
x2_deformed = x₂ + u2
```

```julia
plot(title="Original Body (Blue) vs Deformed Body (Red)", xlabel = "x₁", ylabel = "x₂", aspect_ratio = 1)

plot!(x₁[:,1],x₂[:,1], label = false, color = :blue, lw = 1)
plot!(x₁[:,end],x₂[:,end], label = false, color = :blue, lw = 1)
plot!(x₁[1,:],x₂[1,:], label = false, color = :blue, lw = 1)
plot!(x₁[end,:],x₂[end,:], label = false, color = :blue, lw = 1)

plot!(x1_deformed[:,1],x2_deformed[:,1], label = false, color = :red, lw = 1)
plot!(x1_deformed[:,end],x2_deformed[:,end], label = false, color = :red, lw = 1)
plot!(x1_deformed[1,:],x2_deformed[1,:], label = false, color = :red, lw = 1)
plot!(x1_deformed[end,:],x2_deformed[end,:], label = false, color = :red, lw = 1)

plot!(x₁[:,5],x₂[:,5], label = false, color = :blue, lw = 1)
plot!(x1_deformed[:,5],x2_deformed[:,5], label = false, color = :red, lw = 1)
plot!(x₁[5,:],x₂[5,:], label = false, color = :blue, lw = 1)
plot!(x1_deformed[5,:],x2_deformed[5,:], label = false, color = :red, lw = 1)
```

# Question 2

```julia
using Plots

r = 1:0.01:2
θ = 0:0.01:2π

rr = repeat(r', length(θ))
θθ = repeat(θ, 1, length(r))

x = rr .* cos.(θθ)
y = rr .* sin.(θθ)

ur = 0.4 .* (rr .- 1).^2 .* cos.(3 .* θθ)
uθ = 0.4 .* (rr .- 1).^3

ux = ur .* cos.(θθ) .- uθ .* sin.(θθ)
uy = ur .* sin.(θθ) .+ uθ .* cos.(θθ)

x_def = x .+ ux
y_def = y .+ uy;
```

```julia
plot(title="Original Body (Blue) vs Deformed Body (Red)", xlabel = "x", ylabel = "y", aspect_ratio = 1)

plot!(x[:,1],y[:,1], label = false, color = :blue, lw = 1)
plot!(x[:,end],y[:,end], label = false, color = :blue, lw = 1)
plot!(x[1,:],y[1,:], label = false, color = :blue, lw = 1)
plot!(x[end,:],y[end,:], label = false, color = :blue, lw = 1)

plot!(x_def[:,1], y_def[:,1], label = false, color = :red, lw = 1)
plot!(x_def[:,end], y_def[:,end], label = false, color = :red, lw = 1)
plot!(x_def[1,:], y_def[1,:], label = false, color = :red, lw = 1)
plot!(x_def[end,:], y_def[end,:], label = false, color = :red, lw = 1)

for i in 1:10:100
    plot!(x[:,i],y[:,i], label = false, color = :blue, lw = 1)
    plot!(x_def[:,i],y_def[:,i], label = false, color = :red, lw = 1)
end

for i in 1:50:600
    plot!(x[i,:],y[i,:], label = false, color = :blue, lw = 1)
    plot!(x_def[i,:],y_def[i,:], label = false, color = :red, lw = 1)
end

plot!()
```

# Question 3

```julia
using Plots

x1 = 0.1:0.1:1
x2 = 0.1:0.1:1

x₁ = repeat(x1', length(x2))
x₂ = repeat(x2, 1, length(x1))

rr = sqrt.((x₁ .^ 2) .+ (x₂ .^ 2))
θθ = atan.(x₂ ./ x₁)

ur = 0.2 .* exp.(x₁)
uθ = 0.2 .* log.(1 .+ x₁ .+ x₂)

u1 = ur .* cos.(θθ) .- uθ .* sin.(θθ)
u2 = ur .* sin.(θθ) .+ uθ .* cos.(θθ)

x1_def = x₁ .+ u1
x2_def = x₂ .+ u2;
```

```julia
plot(title="Original Body (Blue) vs Deformed Body (Red)", xlabel = "x₁", ylabel = "x₂", aspect_ratio = 1)

plot!(x₁[:,1],x₂[:,1], label = false, color = :blue, lw = 1)
plot!(x₁[:,end],x₂[:,end], label = false, color = :blue, lw = 1)
plot!(x₁[1,:],x₂[1,:], label = false, color = :blue, lw = 1)
plot!(x₁[end,:],x₂[end,:], label = false, color = :blue, lw = 1)

plot!(x1_def[:,1],x2_def[:,1], label = false, color = :red, lw = 1)
plot!(x1_def[:,end],x2_def[:,end], label = false, color = :red, lw = 1)
plot!(x1_def[1,:],x2_def[1,:], label = false, color = :red, lw = 1)
plot!(x1_def[end,:],x2_def[end,:], label = false, color = :red, lw = 1)

plot!(x₁[:,5],x₂[:,5], label = false, color = :blue, lw = 1)
plot!(x1_def[:,5],x2_def[:,5], label = false, color = :red, lw = 1)
plot!(x₁[5,:],x₂[5,:], label = false, color = :blue, lw = 1)
plot!(x1_def[5,:],x2_def[5,:], label = false, color = :red, lw = 1)
```

# Qustion 4

```julia
L = 1
H = 0.4
D = 2e-3
R = 75e-3
h = 0.05
```

```julia
import Gmsh: gmsh

gmsh.initialize()
gmsh.model.add("Plate with a Hole")

p1 = gmsh.model.geo.addPoint(0,0,0,h)
p2 = gmsh.model.geo.addPoint(L,0,0,h)
p3 = gmsh.model.geo.addPoint(L,H,0,h)
p4 = gmsh.model.geo.addPoint(0,H,0,h)

p5 = gmsh.model.geo.addPoint(L/2,H/2,0,h)
p6 = gmsh.model.geo.addPoint((L/2)+R, H/2,0,h)
p7 = gmsh.model.geo.addPoint((L/2)-R, H/2,0,h)
p8 = gmsh.model.geo.addPoint((L/2), (H/2)+R,0,h)
p9 = gmsh.model.geo.addPoint((L/2), (H/2)-R,0,h)


l1 = gmsh.model.geo.addLine(p1,p2)
l2 = gmsh.model.geo.addLine(p2,p3)
l3 = gmsh.model.geo.addLine(p3,p4)
l4 = gmsh.model.geo.addLine(p4,p1)

l5 = gmsh.model.geo.addCircleArc(p8,p5,p6)
l6 = gmsh.model.geo.addCircleArc(p6,p5,p9)
l7 = gmsh.model.geo.addCircleArc(p9,p5,p7)
l8 = gmsh.model.geo.addCircleArc(p7,p5,p8)

plate_loop = gmsh.model.geo.addCurveLoop([l1,l2,l3,l4])
hole_loop = gmsh.model.geo.addCurveLoop([l5,l6,l7,l8])

surface = gmsh.model.geo.addPlaneSurface([plate_loop, hole_loop])
gmsh.model.geo.synchronize()

ext = gmsh.model.geo.extrude([(2, surface)], 0,0,D)

gmsh.model.addPhysicalGroup(3, [ext[1][2]], 1)
gmsh.model.setPhysicalName(3, 1, "PlateVolume")

gmsh.model.addPhysicalGroup(1, [l1], 2)
gmsh.model.setPhysicalName(1, 2, "LeftBoundary")

gmsh.model.addPhysicalGroup(1, [l2], 3)
gmsh.model.setPhysicalName(1, 3, "RightBoundary")

gmsh.model.geo.synchronize()
gmsh.model.mesh.generate(3)

gmsh.write("plate_with_hole.msh")
gmsh.fltk.run()
gmsh.finalize()
```

## Governing PDE's

∇⋅σ = 0 (Equillibrium)

ϵ = (1/2)((∇⋅u) + (∇⋅u)ᵀ)

σ = 2μϵ + λ(∇⋅u)I (Lame's equation) where μ and λ depend on E and ν

u(left edge) = 0 (Fixed Boundary)   
T(upper edge) = T(lower edge) = T(inner hole surface) =0 (Tractionless surfaces)
